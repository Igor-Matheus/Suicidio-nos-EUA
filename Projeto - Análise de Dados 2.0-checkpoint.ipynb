```python
# Criar um DataFrame contendo apenas as informações relacionadas ao país de interesse: Estados Unidos
# Extrair dados relativos aos EUA
import pandas as pd

df = pd.read_csv("suicidios.csv")

df.rename(columns={'population': 'População'}, inplace = True)
df.rename(columns={'suicides/100k pop': 'Número de suicídios por 100 mil habitantes'}, inplace = True)
df.rename(columns={'gdp_per_capita ($)': 'Produto Interno Bruto (PIB) per capita'}, inplace = True)

df_eua = df[df.country == "United States"].copy()

# Ver o formato do novo Dataframe, as cinco primeiras entradas e as informações

print("Formato do Dataframe: ", df_eua.shape)

df_eua.head()
```

```python
""""
Analisando a tendência da taxa de suicídio nos EUA:

Comparar média do número de suicídios por 100 mil habitantes cometidos nos EUA e a tendência em relação à taxa mundial.

"""
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(9,6))
plt.rcParams.update({'font.size':14})

# Obter a média mundial e a dos EUA em suicídios

anos = df_eua.year.unique()  # pega os anos para o eixo x

media_suicidios_eua = df_eua.groupby('year')['Número de suicídios por 100 mil habitantes'].mean()
media_suicidios_mundo = df.groupby('year')['Número de suicídios por 100 mil habitantes'].mean()

# Como os EUA nao tem 2016, vou eliminar do dataframe mundial essa entrada

media_suicidios_mundo.drop(2016, inplace=True)

# Gráfico comparativo entre EUA e o Mundo

ax = sns.lineplot(x = anos, y = media_suicidios_eua, label = 'Estados Unidos')
ax = sns.lineplot(x = anos, y = media_suicidios_mundo, label = 'Mundo')

plt.legend(title = "Média de suicídios")

plt.xlabel('Ano')

plt.ylabel('Número de suicídios por 100 mil habitantes')

plt.savefig('Gráfico 1.png', dpi = 600) # salva o gráfico em formato png

plt.show()
```

```python
"""
Analisando a taxa de suicídio entre homens e mulheres nos EUA:

Pegar a média dos 30 anos, pois não houve mudança significativa desse comportamento durante o período.

"""

# Obter valores entre Homens e Mulheres

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("suicidios.csv")
df_eua = df[df.country == "United States"].copy()

sexo_suicidios = df_eua.groupby('sex').suicides_no.sum() / df_eua.groupby('sex').suicides_no.sum().sum()

# Gráfico mostrando porcentagem de homens e mulheres 

plt.pie(sexo_suicidios, labels = ['Mulheres', 'Homens'], autopct = '%1.1f%%', shadow = True);
```

```python
"""
Faixa etária com maior índice de suícidio:

Abaixo, uma tabela dinâmica (pivot table) para analisar as 6 faixas etárias em função do ano e do número de suicídios.

O principal objetivo aqui é identificar qual a faixa etária que tem a maior representatividade entre aqueles que tiram a
própria vida, e identificar se houve alguma mudança no perfil ao longo de 30 anos.

"""

# Criar uma tabela dinâmica

tabela = pd.pivot_table(df_eua, values='suicides_no', index=['year'], columns=['age'])

# Reordenar as tableas para deixar em ordem crescente

column_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']
tabela = tabela.reindex(column_order, axis=1)

# Mudar idade para português
tabela.rename(columns={'5-14 years': '5-14 anos'}, inplace = True)
tabela.rename(columns={'15-24 years': '15-24 anos'}, inplace = True)
tabela.rename(columns={'25-34 years': '25-34 anos'}, inplace = True)
tabela.rename(columns={'35-54 years': '35-54 anos'}, inplace = True)
tabela.rename(columns={'55-74 years': '55-74 anos'}, inplace = True)
tabela.rename(columns={'75+ years': '75+ anos'}, inplace = True)

# Ver as 5 primeiras entradas da pivot table

tabela.head() 
```

```python
# Obter a pivot table das faixas etárias

plt.rcParams.update({'font.size':14})

tabela.plot.bar(stacked = 1, figsize = (13,6),
                title = 'Média do número de suicídios entre diferentes faixas etárias por ano nos EUA')

plt.legend (title = "Idade", bbox_to_anchor = (1.04,1), loc = "best")

plt.xlabel('Ano')

plt.savefig('Gráfico 4.png', dpi = 600) # salva o gráfico em formato png
```

```python
"""
Correlações entre o PIB percapita e número de suicídios por 100 mil habitantes:

Criando uma matriz de correlação e plotando um heatmap, infere-se que o aumento no PIB per capita não diminuiu o número
de suicídios por 100 mil habitantes. Na verdade, ele se manteve estável, contrariando o senso comum da maioria das pessoas.

Em relação ao IDH, há muitos valores ausentes nas células, o que pode dar uma interpretação incorreta ou com viés.

"""

# Elaborar um heatmap para as correlações

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("suicidios.csv")

df.rename(columns={'population': 'População'}, inplace = True)
df.rename(columns={'suicides/100k pop': 'N° de suicídios por 100 mil hab'}, inplace = True)
df.rename(columns={'gdp_per_capita ($)': 'PIB per capita'}, inplace = True)

# Tirar do heatmap o que não vai nos interessar nele
# O IDH foi tirado porque tinham muitos dados vazios sobre ele

df.drop('HDI for year', axis = 1, inplace = True)
df.drop('year', axis = 1, inplace = True)
df.drop('suicides_no', axis = 1, inplace = True)

df.head()
```

```python
plt.rcParams.update({'font.size':12})

df_eua = df[df.country == "United States"].copy()

sns.heatmap(df_eua.corr(), cmap = 'Wistia', annot = True)

plt.savefig('Gráfico 6.png', dpi = 600) # salva o gráfico em formato png
```

```python
"""
Gráfico do tipo donut mostrando as porcentagens por gênero

"""

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(5,7))
plt.rcParams.update({'font.size':14})

df = pd.read_csv("suicidios.csv")
df_eua = df[df.country == "United States"].copy()

numero_homens_mulheres = df_eua.groupby('sex').suicides_no.sum()

sum_sum = df_eua.groupby('sex').suicides_no.sum().sum()

print(" ", numero_homens_mulheres) # Número de suicídios de homens e número de suicídios de mulheres durante o período.

print("\nTotal: ", sum_sum) # Número total de suicídios durante o período.

sexo_suicidios = df_eua.groupby('sex').suicides_no.sum() / df_eua.groupby('sex').suicides_no.sum().sum()

plt.pie(sexo_suicidios, labels = ['Mulheres', 'Homens'],  wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' },
        autopct = '%1.1f%%', shadow = 0, colors = ('aquamarine', 'cornflowerblue'), labeldistance = 1.05);

circulo1 = plt.Circle( (0,0), 0.75, color = 'white')
p = plt.gcf()
p.gca().add_artist(circulo1)

plt.savefig('Gráfico 2.png', dpi = 600) # salva o gráfico em formato png

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size':14})

sexo = ['Período: 1985 a 2015']

numerosM = [213797]
numerosH = [820216]

x = np.arange(len(sexo))

largura = 0.16

fig,ax = plt.subplots(figsize = (10,6))

grupo1 = ax.bar(x - largura/2, numerosM, largura, label = 'Mulheres', color = 'aquamarine')
grupo2 = ax.bar(x + largura/2, numerosH, largura, label = 'Homens', color = 'cornflowerblue')

ax.set_title('Número de suicídios nos EUA por gênero')

ax.legend()

ax.set_ylim([0,980000])

ax.set_xticks(x)

ax.set_xlim([-0.5,0.5])

ax.set_xticklabels(sexo)

def autolabel(grupo):
    for i in grupo:
        h = i.get_height()
        ax.annotate('{}'.format(h),
                xy = (i.get_x()+i.get_width()/2,h),
                xytext = (0,3),
                textcoords = 'offset points',
                ha = 'center')
                    
autolabel(grupo1)  
autolabel(grupo2)

plt.savefig('Gráfico 3.png', dpi = 600) # salva o gráfico em formato png
                    
plt.show()
```

```python
# Histograma de homens e mulheres ao longo dos anos, se possível:

# Usar uma pivot table

import pandas as pd

df = pd.read_csv("suicidios.csv")
df_eua = df[df.country == "United States"].copy()

tabela2 = pd.pivot_table(df_eua, values='suicides_no', index=['year'], columns=['sex'])

tabela2.rename(columns={'female': 'Mulheres'}, inplace = True)
tabela2.rename(columns={'male': 'Homens'}, inplace = True)

tabela2.head()
```

```python
# Histograma a partir da pivot table criada

plt.rcParams.update({'font.size':14})

column_order2 = ['Homens', 'Mulheres']
tabela2 = tabela2.reindex(column_order2, axis=1)

tabela2.plot.bar(stacked = 1, figsize = (13,6), title = 'Média do número de suicídios entre homens e mulheres por ano nos EUA',
                 color = {"Mulheres": "aquamarine", "Homens": "cornflowerblue"})

plt.legend (title = "Sexo", bbox_to_anchor = (1.04,1), loc = "best")
plt.xlabel('Ano')
```
